## Figure 4 · Convergence speed per optimizer

```python
thresholds = [0.70, 0.75, 0.80, 0.85, 0.90, 0.95]
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, mn, lbl in zip(axes, ['deit_tiny','resnet18'], ['DeiT-Tiny','ResNet-18']):
    for r in get_by_model(mn):
        ep_needed = []
        for thr in thresholds:
            ep = next((i+1 for i,a in enumerate(r['history']['val_acc']) if a>=thr), None)
            ep_needed.append(ep if ep else 21)
        ax.plot(thresholds, ep_needed, marker='o', label=r['optimizer'],
                color=COLORS[r['optimizer']], linewidth=1.8)
    ax.set_xlabel('Accuracy Threshold', fontsize=10)
    ax.set_ylabel('Epochs needed', fontsize=10)
    ax.set_title(f'{lbl} — Epochs to reach threshold', fontweight='bold')
    ax.legend(fontsize=7); ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 22); ax.set_yticks(range(0, 23, 2))
    ax.axhline(y=21, color='gray', linestyle='--', alpha=0.5, label='Not reached')
plt.tight_layout()
plt.show()
print('Saved convergence_speed.png')
```

## Figure 5 · Train vs Validation accuracy gap (Overfitting analysis)

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, mn, lbl in zip(axes, ['deit_tiny','resnet18'], ['DeiT-Tiny','ResNet-18']):
    for r in get_by_model(mn):
        opt = r['optimizer']
        gap = [tr - va for tr, va in zip(r['history']['train_acc'], r['history']['val_acc'])]
        ax.plot(range(1, 21), gap, label=opt, color=COLORS[opt], linewidth=1.8)
    ax.axhline(y=0, color='black', linestyle='--', alpha=0.4)
    ax.set_xlabel('Epoch'); ax.set_ylabel('Train Acc - Val Acc')
    ax.set_title(f'{lbl} — Overfitting Gap', fontweight='bold')
    ax.legend(fontsize=7); ax.grid(True, alpha=0.3)
    ax.xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
plt.tight_layout()
plt.show()
print('Saved overfitting_gap.png')
```
